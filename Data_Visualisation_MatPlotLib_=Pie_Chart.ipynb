#__PieCharts__
a pie chart is a circular statistical graphic that displays data as slices of a pie. Each slice represents a category of data, and the size of the slice is proportional to the category's contribution to the whole.

```python
import matplotlib.pyplot as plt #Importing Matplotlib library
import numpy as np #Importing Numpy library

y = np.array([10, 20, 30, 40])

plt.pie(y) #Creating Pie Chart by using y varibale values.
plt.show()
```

###__Labels__
Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge.

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])

plt.pie(y, labels=labels) #Using labels and then creating the Pie Chart
plt.show()
```

###__Start Angle__
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0.

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])

plt.pie(y, labels=labels,  startangle = 90) #Using startangle to start angles
plt.show()
```

###__Explode__
The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed.

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])
explode = np.array([0.3,0,0,0]) #Here the specific explode is given to assign indexes means, ten=0.3, twenty,thirty,fourty=0.

plt.pie(y, labels=labels,  startangle = 90, explode = explode) #Using explode parameter we can seperate item.
plt.show()
```

###__Shadow__
We can add a shadow to the pie chart by setting the shadows parameter to True.

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])
explode = np.array([0.3,0,0,0])

plt.pie(y, labels=labels,  startangle = 90, explode = explode, shadow=True) #Adding shadow to Pie chart using shadow parameter.
plt.show()
```

###__Colors__
We can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge.

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])
colors = np.array(['red','blue','orange','pink'])
explode = np.array([0.3,0,0,0])

plt.pie(y, labels=labels,  startangle = 90, explode = explode, shadow=True, colors = colors) #Adding colors to Pie chart using color parameter.
plt.show()
```

###__Legend__
To add a list of explanation for each wedge, use the legend() function

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])
colors = np.array(['red','blue','orange','pink'])
explode = np.array([0.3,0,0,0])

plt.pie(y, labels=labels,  startangle = 90, explode = explode, shadow=True, colors = colors)
plt.legend() #Adding legend to Pie Chart for better understanding.
plt.show()
```

###__Legend With Header__
To add a header to the legend, add the title parameter to the legend function.

```python
y = np.array([10,20,30,40])
labels = np.array(['Ten','Twenty','Thirty','Fourteen'])
colors = np.array(['red','blue','orange','pink'])
explode = np.array([0.3,0,0,0])

plt.pie(y, labels=labels,  startangle = 90, explode = explode, shadow=True, colors = colors)
plt.legend(title='Legend Header') #Adding title to legend .
plt.show()
```
