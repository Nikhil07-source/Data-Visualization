#__Matplotlib Bars__
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

```python
import matplotlib.pyplot as plt #Importing matplotlib Library with pyplot module
import numpy as np #Importing numpy library

subject = ['Math','Science','S.St','English','Hindi','Computer','Sanskrit'] #Creating a list where subjects are as list items
marks = [42,65,55,78,89,98,62] #Creating a list where marks are as list items

plt.bar(subject, marks) #Here we are creating a Bar Graph assigning subject to x axis and marks as y axis.

plt.title('Student Marks') #This is used to give title to our graph
plt.xlabel('Subjects') #This is used to assign name to x axis
plt.ylabel('Marks') #This is used to assign name to y axis
plt.show() #This is used to show the output.
```

###__Horizontal Bars__
If we want the bars to be displayed horizontally instead of vertically, use the barh() function.

```python
subject = ['Math','Science','S.St','English','Hindi','Computer','Sanskrit']
marks = [42,65,55,78,89,98,62]

plt.barh(subject, marks) #Here we are using plt.barh, using 'h' infront of plt.bar we are able to create Horizontal Bar Graph.

plt.title('Student Marks as Horizontal ')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.show()
```

###__Bar Height__
The barh() takes the keyword argument height to set the height of the bars.

```python
subject = ['Math','Science','S.St','English','Hindi','Computer','Sanskrit']
marks = [42,65,55,78,89,98,62]

plt.barh(subject, marks, height = 0.1) #Here we are setting the height of the bar

plt.title('Student Marks as Horizontal ')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.show()
```

###__Bar Width__
The bar() takes the keyword argument width to set the width of the bars

```python
subject = ['Math','Science','S.St','English','Hindi','Computer','Sanskrit']
marks = [42,65,55,78,89,98,62]

plt.bar(subject, marks, color='red', width = 0.1) #Here we giving width to the bars.

plt.title('Student Marks as Horizontal ')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

###__Bar Color__
The bar() and barh() take the keyword argument color to set the color of the bars

```python
subject = ['Math','Science','S.St','English','Hindi','Computer','Sanskrit']
marks = [42,65,55,78,89,98,62]

plt.bar(subject, marks, color='red') #Here we are color to set the color of the bar. We can use text, color code to set color.

plt.title('Student Marks as Horizontal ')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```
